1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
def most_viewed_video_youtube():
    #intiating the browser
    driver=webdriver.Chrome('C:/Users/yn/Desktop/Yuvi/DataTrained/seleniumWebdriver/chromedriver.exe')
    #loading the given url
    driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
    time.sleep(3)
    table_body_tr_list=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr')
    table_head_tr=driver.find_element_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/thead/tr')
    head_childrens=table_head_tr.find_elements_by_xpath('//th')
    
    rank_pos=table_head_tr.find_element_by_xpath('//th[abbr[.="No."]]')
    name_pos=table_head_tr.find_element_by_xpath('//th[contains(.,"Video name")]')
    artist_pos=table_head_tr.find_element_by_xpath('//th[contains(.,"Uploader")]')
    up_date_pos=table_head_tr.find_element_by_xpath('//th[contains(.,"Upload date")]')
    views_pos=table_head_tr.find_element_by_xpath('//th[contains(.,"Views (billions)")]')
    
    rank_list=[]
    name_list=[]
    artist_list=[]
    up_date_list=[]
    view_list=[]
    
    for tr in table_body_tr_list:
        try:
            rank_list.append(tr.find_element_by_xpath('.//td[position()='+str(head_childrens.index(rank_pos)+1)+']').text)
        except:
            rank_list.append('-')
        try:
            name=tr.find_element_by_xpath('.//td[position()='+str(head_childrens.index(name_pos)+1)+']').text.replace('"','')
            name_list.append(name[:name.index('[')])
        except:
            name_list.append('-')
        try:
            artist_list.append(tr.find_element_by_xpath('.//td[position()='+str(head_childrens.index(artist_pos)+1)+']/a').text)
        except:
            artist_list.append('-')            
        try:
            up_date_list.append(tr.find_element_by_xpath('.//td[position()='+str(head_childrens.index(up_date_pos)+1)+']').text)
        except:
            up_date_list.append('-')
        try:
            view_list.append(tr.find_element_by_xpath('.//td[position()='+str(head_childrens.index(views_pos)+1)+']').text)
        except:
            view_list.append('-')
    res_table=pd.DataFrame({'Rank':rank_list,'Name':name_list,'Artist':artist_list,'Uploaded Date':up_date_list,'Views':view_list})      
    return res_table
```

5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
def github():
    driver=webdriver.Chrome('C:/Users/yn/Desktop/Yuvi/DataTrained/seleniumWebdriver/chromedriver.exe')
    #loading the given url
    driver.get('https://github.com/')
    time.sleep(10)
    driver.maximize_window()
    repo_title=[]
    repo_desc=[]
    contrib_count=[]
    href_list=[]
    lang_list=[]
    
    element_to_hover_over = driver.find_element_by_xpath('//summary[contains(.,"Explore")]')
    hover = ActionChains(driver).move_to_element(element_to_hover_over)
    hover.perform()
    #driver.find_element_by_xpath('//summary[contains(.,"Explore")]').click()
    #time.sleep(10)
    #driver.find_element_by_xpath('//a[contains(@href,"explore")]').click()
    try:
        href=driver.find_element_by_xpath('//a[contains(@href,"trending")]').get_attribute('href')
    except:
        hover = ActionChains(driver).move_to_element(element_to_hover_over)
        hover.perform()
        time.sleep(5)
        driver.find_element_by_xpath('//a[contains(@href,"trending")]').click()
    driver.get(href)
    #rticle_list=driver.find_elements_by_xpath('//div[@class="Box"]/div[2]/article')                 
    for hr in driver.find_elements_by_xpath('//div[@class="Box"]/div[2]/article//h1/a'):
        href_list.append(hr.get_attribute('href'))
    for href in href_list:
        lang_str=''
        driver.execute_script("window.open('');")
        driver.switch_to.window(driver.window_handles[1])
        driver.get(href)
        time.sleep(10)
        try:
            repo_title.append(driver.find_element_by_xpath('//strong[@itemprop="name"]').text)
        except:
            repo_title.append('-')  
        try:
            contrib_count.append(driver.find_element_by_xpath('//a[contains(@href,"contributor")]/span[@class="Counter"]').text)
        except:
            contrib_count.append('-') 
        try:
            repo_desc.append(driver.find_element_by_xpath('//h2[.="About"]/following-sibling::p[1]').text)
        except:
            repo_desc.append('-')
        try:
            for i in driver.find_elements_by_xpath('//h2[.="Languages"]/following-sibling::ul[1]/li//span[contains(@class,"default text-bold")]'):
                lang_str+=i.text+','
            lang_list.append(lang_str[:-1])
        except:
            lang_list.append('-')                
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
    
    res_table=pd.DataFrame({'Repository title':repo_title,'Repository description':repo_desc,
                            'Contributors count':contrib_count,
                           "Language used":lang_list})
    return res_table
     
        
    
```

```python
def icsuci():
    driver=webdriver.Chrome('C:/Users/yn/Desktop/Yuvi/DataTrained/seleniumWebdriver/chromedriver.exe')
    #loading the given url
    driver.get('https://archive.ics.uci.edu/')
    time.sleep(15)
    driver.maximize_window()
    driver.find_element_by_xpath('//a[.="View ALL Data Sets"]').click()
    time.sleep(10)
    
    data_list=[]
    data_type_list=[]
    task_list=[]
    attrib_type_list=[]
    instance_list=[]
    attribute_list=[]
    year_list=[]
    
    data_sets_list=driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr[td/p[@class="normal"]]')
    for data in data_sets_list:
        try:
            data_list.append(data.find_element_by_xpath('./td[1]//td[2]').text)
        except:
            data_list.append('-')
        try:
            data_type_list.append(data.find_element_by_xpath('./td[2]').text)
        except:
            data_type_list.append('-')
        try:
            task_list.append(data.find_element_by_xpath('./td[3]').text)
        except:
            task_list.append('-')
        try:
            attrib_type_list.append(data.find_element_by_xpath('./td[4]').text)
        except:
            attrib_type_list.append('-')
        try:
            instance_list.append(data.find_element_by_xpath('./td[5]').text)
        except:
            instance_list.append('-')
        try:
            attribute_list.append(data.find_element_by_xpath('./td[6]').text)
        except:
            attribute_list.append('-')
        try:
            year_list.append(data.find_element_by_xpath('./td[7]').text)
        except:
            year_list.append('-')
    res_table=pd.DataFrame({'Dataset name':data_list,'Data type':data_type_list,
                           'Task':task_list,'Attribute type':attrib_type_list,'No of instances':instance_list,
                           'No of attribute':attribute_list,'Year':year_list})
    return res_table
```
